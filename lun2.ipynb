```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import os


chrome_options = Options()
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36")

# Set the path to the chromedriver, change for the windows path
driver_path = './chromedriver_112.0.5615.49' if os.name == 'posix' else './chromedriver.exe'

service = Service(executable_path=driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# Navigate to the website
driver.get('https://lun.ua/uk/%D0%BD%D0%B5%D1%80%D1%83%D1%85%D0%BE%D0%BC%D1%96%D1%81%D1%82%D1%8C-%D0%BA%D0%B8%D1%94%D0%B2%D0%B0')

# Simulate user actions
time.sleep(5) # wait for the page to load
# use this for user actions and page changing

# search_box = driver.find_element_by_name('q')
# search_box.send_keys('search query')
# search_box.submit()

# Extract data
soup = BeautifulSoup(driver.page_source, 'html.parser')
results = soup.find_all(id='search-results')

print(results)

# Close the browser
driver.quit()
```
